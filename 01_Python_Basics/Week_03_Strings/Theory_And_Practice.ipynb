1. Definition
A string is a sequence of characters (letters, numbers, symbols, spaces) enclosed in quotes.

name = "Ayush"              # String
message = 'Hello World'     # String
empty = ""                  # Empty string

Key Point: In Python, strings are immutable (you can't change them after creation, but you can create new ones).

```python
# 3.1 Creating Strings

single_quote = 'Hello'
double_quote = "World"
triple_quote = '''Multi
line
string'''

# Escape characters
escaped = "He said \"Hello\""  # " inside string
newline = "Line 1\nLine 2"     # \n = new line
tab = "Name\tAge"              # \t = tab
print(escaped)
print(newline)
print(tab)
```

```python
# 3.2 String Methods (Most Important)

text = "hello world"

# Case conversion
text.upper()                    # "HELLO WORLD"
text.lower()                    # "hello world"
text.capitalize()               # "Hello world"
text.title()                    # "Hello World"

# Finding & Replacing
text.find("world")              # 6 (index where "world" starts)
text.count("l")                 # 3 (count of "l")
text.replace("world", "Python") # "hello Python"

# Splitting & Joining
words = text.split()            # ["hello", "world"]
" ".join(words)                 # "hello world"

# Checking
text.startswith("hello")        # True
text.endswith("world")          # True
text.isdigit()                  # False
text.isalpha()                  # False (has space)

# Stripping whitespace
"  hello  ".strip()             # "hello" (removes leading/trailing spaces)
```

```python
# 3.3 String Indexing & Slicing

text = "Python"

text[0]              # "P" (first character)
text[-1]             # "n" (last character)
text[0:3]            # "Pyt" (index 0,1,2)
text[1:]             # "ython" (from index 1 to end)
text[:3]             # "Pyt" (from start to index 3)
text[::2]            # "Pto" (every 2nd character)
```

```python
# 3.4 String Formatting
name = "Ayush"
age = 20

# Old way (not recommended)
"Name: %s, Age: %d" % (name, age)

# Better way
"Name: {}, Age: {}".format(name, age)

# BEST way (f-strings)
f"Name: {name}, Age: {age}"
```

6. Beginner Mistakes

❌ Mistake 1: Forgetting Strings Are Immutable

# WRONG
text = "hello"
text[0] = "H"  # ❌ TypeError: strings don't support item assignment


```python
# CORRECT
text = "hello"
text = "H" + text[1:]  # Create new string
print(text)  # "Hello"
```

❌ Mistake 2: Confusing in with Index
text = "hello"

# WRONG
text.find("e")  # Returns 1 (index)
if text.find("e"):
    print("Found")  # This ALWAYS prints (even if "e" at index 0)

```python
# CORRECT
text = "hello"
if "e" in text:
    print("Found")  # Only prints if "e" exists
```

❌ Mistake 3: Forgetting join() Works Backwards

# WRONG
list_of_words = ["apple", "banana", "cherry"]
result = list_of_words.join("-")  # ❌ AttributeError


```python
# CORRECT
list_of_words = ["apple" , "banana" ,"cherry"]
result = " + ".join(list_of_words)  # "apple-banana-cherry"
print(result)
```

❌ Mistake 4: Not Stripping Input

# WRONG (from user input)
name = input("Enter name: ")  # User types: "  Ayush  \n"
if name == "Ayush":
    print("Found")  # Won't match because of spaces!

NOTE- The computer treats spaces as part of the string.


```python
# CORRECT
name = input("Enter name: ").strip()
if name == "Ayush":
    print("Found")  # Now it matches
else:
    print("Not found")
```

❌ Mistake 5: Forgetting String Index Errors

# WRONG
text = "Hi"

```python
# CORRECT
text = "Hi"
if len(text) > 10:
    print(text[10])
else:
    print("Index not found")
```

8. Important Notes

⚠️ Strings vs Lists:

# String (immutable)
text = "hello"
text[0] = "H"  # ❌ Can't change

# List (mutable)
chars = list("hello")  # ['h','e','l','l','o']
chars[0] = "H"         # ✅ Can change
text = "".join(chars)  # Convert back to string
----------------------------------------------------

⚠️ Raw Strings (for file paths):

# WRONG
path = "C:\new\folder"  # ❌ \n and \f are escape sequences

# CORRECT
path = r"C:\new\folder"  # Raw string (no escaping)
-------------------------------------------------------

⚠️ String Multiplication:

print("Ha" * 3)  # "HaHaHa"
print("-" * 10)  # "----------"
---------------------------------------------------------

⚠️ Empty String Check:

text = ""
if text:        # False (empty string is falsy)
    print("Has text")
if not text:    # True
    print("No text")

```python
#solution H1

def is_palindrome(text):
    # Ignore whitespace
    text = text.replace(" " , "")

    # Convert to lowercase
    text = text.lower()

    # Check if the string is the same forwards and backwards
    return text == text[::-1]

# Test with different inputs
print(is_palindrome("maDam"))
print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
```

```python
#Solution H2

def compress_string(text):
    # If the string is empty, return an empty string
    if text == "":
        return ""

    # Store the compressed result
    compressed = ""

    # Count of the current character
    count = 1

    # Loop through the string starting from the second character
    for i in range(1, len(text)):

        # If current character is the same as previous character
        if text[i] == text[i - 1]:
            count += 1

        # If character changes
        else:
            # Add previous character and its count
            compressed += text[i - 1] + str(count)

            # Reset count for the new character
            count = 1

    # Add the last character and its count
    compressed += text[-1] + str(count)

    # If compressed string is not shorter,
    # return the original string
    if len(compressed) >= len(text):
        return text
    else:
        return compressed



# Test Cases


print(compress_string("aabbbcccc"))    # a2b3c4
print(compress_string("abcdef"))       # abcdef
print(compress_string("aa"))           # aa
print(compress_string("aaaaa"))        # a5
print(compress_string("aabb"))         # aabb
print(compress_string("aaabccdddd"))   # a3b1c2d4
print(compress_string(""))             # (empty string)
```

12. Revision Notes

Key Takeaways:

1. Strings are sequences of characters indexed from 0.
2. Strings are immutable — you create new strings, not modify existing ones.
3. Indexing: text[0] is first char, text[-1] is last.
4. Slicing: text[start:end:step] — remember end is exclusive.
5. Common methods: upper(), lower(), split(), strip(), replace(), find().
6. F-strings are best: f"{variable}".
7. .join() works with sequences: "-".join(list_of_strings).
8. Check membership: "e" in text (not text.find("e")).
9. Raw strings for paths: r"C:\path\folder".
10. Always .strip() user input: input().strip().
