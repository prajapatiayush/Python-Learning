Q2: Fix this code (explain the error and fix it):
email = input("Enter email: ")
if email == "ayush@example.com":
    print("Welcome")

# Problem: User types "  ayush@example.com  \n"
-----------------------------------------------------------------------

- Why does it fail?

If the user enters:

"  ayush@example.com  "

The value stored in email is:

"  ayush@example.com  "

This is not equal to:

"ayush@example.com"

because of the leading and trailing spaces.
----------------------------------------------------------------------------------------------

- Explanation:

1. input() reads the entire user input as a string.
2..strip() removes:
   Leading spaces
   Trailing spaces
   Tabs (\t)
   Newline characters (\n)

```python
# Fixed Code 

email = input("Enter email: ").strip()

if email == "ayush@example.com":
    print("Welcome ! ")
else:
    print("Invalid email")
```

Q3: Count word frequency in a sentence:
sentence = "the quick brown fox jumps over the lazy dog the"
words = sentence.split()
# Count how many times each word appears
# Output: {"the": 3, "quick": 1, "brown": 1, ...}

```python
# Solution
sentence = "the quick brown fox jumps over the lazy dog the"

# Split the sentence into words
words = sentence.split()

# Create an empty dictionary
word_count = {}

# Count the frequency of each word
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Display the result
print(word_count)
```

Q4: Check if a string is a palindrome (ignore case and spaces):
text = "A man a plan a canal Panama"
# Return True if palindrome, False otherwise
# (Hint: remove spaces, convert to lowercase, reverse and compare)

```python
#Solution
text = "A man a plan a canal Panama"

# Remove spaces and convert to lowercase
cleaned_text = text.replace(" ", "").lower()

# Reverse the string
reversed_text = cleaned_text[::-1]

# Compare both strings
if cleaned_text == reversed_text:
    print(True)
else:
    print(False)
```

Q5: Parse a stock ticker string:
ticker = "RELIANCE.NS:₹2500:5M"
# Extract: Company, Exchange, Price, Volume
# Output:
# Company: RELIANCE
# Exchange: NS
# Price: ₹2500
# Volume: 5M

```python
# Solution
ticker = "RELIANCE.NS:₹2500:5M"

# Split using ':'
parts = ticker.split(":")

# Split company and exchange using '.'
company, exchange = parts[0].split(".")

# Extract remaining values
price = parts[1]
volume = parts[2]

# Display the result
print("Company :", company)
print("Exchange:", exchange)
print("Price   :", price)
print("Volume  :", volume)
```
