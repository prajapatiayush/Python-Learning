Project: Stock Price Alert Message Generator
Create a program that generates professional trading alert messages.

Input:
company_name = "RELIANCE"
current_price = 2500
previous_close = 2400
percentage_change = ((current_price - previous_close) / previous_close) * 100
volume = 5000000
alert_type = "bullish"  # "bullish" or "bearish"
-------------------------------------------------------------------

Features:
Format company name properly (capitalize)
Calculate and format percentage change
Create formatted alert message
Validate inputs
Generate timestamp
-------------------------------------------------------------------

Expected Output:
═══════════════════════════════════════════
📊 TRADING ALERT - RELIANCE
═══════════════════════════════════════════
Price: ₹2,500
Change: +4.17% ⬆️ (from ₹2,400)
Volume: 5,000,000
Alert Type: BULLISH

Message: RELIANCE has surged 4.17% to ₹2,500 on strong volume!
═══════════════════════════════════════════

Code Requirements:
Use string formatting (f-strings)
Use .format() or similar for currency
Use string methods (upper, lower, replace, etc.)
Validate that percentage is numeric
Add timestamp

```python
# Stock Price Alert Message Generator

from datetime import datetime

# Input
company_name = "RELIANCE"
current_price = 2500
previous_close = 2400
volume = 5000000
alert_type = "bullish"      # bullish or bearish

# Calculate percentage change
percentage_change = ((current_price - previous_close) / previous_close) * 100

# Validate inputs
if current_price <= 0 or previous_close <= 0:
    print("Invalid price!")
else:

    # Format company name
    company = company_name.upper()

    # Format currency
    current_price_str = "₹{:,.0f}".format(current_price)
    previous_close_str = "₹{:,.0f}".format(previous_close)

    # Format percentage
    percent = "{:.2f}%".format(percentage_change)

    # Format volume
    volume_str = "{:,}".format(volume)

    # Generate timestamp
    timestamp = datetime.now().strftime("%d-%m-%Y %H:%M:%S")

    # Create message
    if alert_type.lower() == "bullish":
        message = f"{company} has surged {percent} to {current_price_str} on strong volume!"
    else:
        message = f"{company} has fallen {percent} to {current_price_str} on heavy selling!"

    # Display Alert
    print("=" * 55)
    print(f"📈 TRADING ALERT - {company}")
    print("=" * 55)

    print("Price:", current_price_str)
    print(f"Change: +{percent} (from {previous_close_str})")
    print("Volume:", volume_str)
    print("Alert Type:", alert_type.upper())
    print("Time:", timestamp)

    print()
    print("Message:", message)

    print("=" * 55)
```
