```python
# Q4 (Medium):
# Check if a number n is between 100 and 200 (inclusive). Write the condition using:

# Method 1: and operator
# Method 2: Chained comparison

num = int(input("enter number"))

#Method 1: and operator
if num >= 100 and num <=200:
    print("Method 1:number is in the range 100-200")
else:
    print("Method 1:number is not in the range")

#Method 2: Chained comparison

if 100 <= num <=200:
    print("Method 2: number is in the range 100-200")
else:
    print("Method 2: number is not in the range")
```
