Project: Loan Eligibility Checker

```python
# Input
age = int(input("Enter age : "))
annual_income = int(input("Enter annual income :"))
credit_score = int(input("Enter credit score : "))
existing_loans = int(input("Enter existing loan : "))

# Check conditions 
age_ok = age >= 21
income_ok = annual_income > 300000
credit_ok = credit_score >= 700
loans_ok = existing_loans <= 3

# Decisions 
eligible = age_ok and income_ok and credit_ok and loans_ok

# Output
print("\n === LOAN ELIGIBILITY RESULT ===")
print(f"Age Check(>= 21): {'Pass'if age_ok else 'Fail'}")
print(f"Income Check(> 300k): {'Pass' if income_ok else 'Fail'}")
print(f"Credit Check(>= 700): {'Pass' if credit_ok else 'Fail'}")
print(f"Existing Loans (<= 3): {'Pass' if loans_ok else 'Fail'}")

if eligible:
    approved = annual_income * 345
    print(f"\n APPROVES FOR LOAN")
    print(f"\n Approved Amount : ₹{approved} ")
else:
    print(f"\n NOT ELIGIBLE")
```
