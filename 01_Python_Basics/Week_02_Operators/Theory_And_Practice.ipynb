1. Definition
Operators are special symbols in Python that perform operations on variables and values.
An operator takes operands (values/variables) and produces a result.

a + b # "+" is an operator , "a" and "b" are operands

Types of Operators:

Arithmetic Operators — Math operations
Comparison Operators — Compare two values
Logical Operators — Boolean logic (True/False)
Assignment Operators — Assign values
Bitwise Operators — Work with binary (Advanced)


2. Why It Is Important

Scenario                 |              Why Needed

Without operators, you can't: Make decisions, do math, build logic, or create algorithms.

3. Syntax Breakdown

-Arithmetic Operators

a + b      # Addition
a - b      # Subtraction
a * b      # Multiplication
a / b      # Division (always returns float)
a // b     # Floor Division (returns int)
a % b      # Modulo (remainder)
a ** b     # Exponentiation (power)

-Comparison Operators

a == b     # Equal to
a != b     # Not equal to
a > b      # Greater than
a < b      # Less than
a >= b     # Greater than or equal to
a <= b     # Less than or equal to

-Logical Operators

a and b    # Both must be True
a or b     # At least one must be True
not a      # Reverses boolean value

-Assignment Operators

a = b      # Assign
a += b     # Add and assign (a = a + b)
a -= b     # Subtract and assign
a *= b     # Multiply and assign
a /= b     # Divide and assign

6. Beginner Mistakes
❌ Mistake 1: Confusing = with ==
# WRONG
if a = 5:  # ❌ Syntax error
    print("five")

Why? = is assignment, == is comparison.

```python
# CORRECT
if a == 5:
    print("five")
```

❌ Mistake 2: Division Always Returns Float

# WRONG (expecting integer)
result = 10 / 3
print(result)  # Output: 3.3333... (NOT 3)


```python
# CORRECT (if you want integer)
result = 10//3
print(result)
```

❌ Mistake 3: Wrong Operator Precedence

# WRONG
result = 2 + 3 * 4  # Student thinks: (2+3)*4 = 20
print(result)  # Output: 14 (NOT 20)
# Because * has higher precedence than +

NOTE- Order: ** > *,/,//,% > +,- > Comparisons > and > or

```python
# CORRECT
result = ( 2 + 3) * 4
print(result)
```

❌ Mistake 4: Using and/or Instead of Comparison Operators

```python
# WRONG
if a > 5 and b > 10:  # ✓ This is correct
    pass

# CONFUSING (work but bad practice)
x = 15
y = 10
if x and y:
    pass
```

❌ Mistake 5: Chaining Comparisons Wrong

NOTE# This ACTUALLY works in Python! But many languages don't support it.

8. Important Notes

⚠️ Order of Evaluation:
result = 10 + 5 * 2  # * before +  →  20 (not 30)
result = (10 + 5) * 2  # Parentheses first  →  30

⚠️ Modulo (%) is Useful:
# Check if even/odd
if num % 2 == 0:
    print("Even")
    
# Check if divisible
if price % 5 == 0:
    print("Price is multiple of 5")

⚠️ Floor Division (//) vs Regular Division (/):
python
10 / 3   →  3.3333...  (float)
10 // 3  →  3          (int, rounds down)
-10 // 3 →  -4         (rounds down, not towards zero!)

⚠️ Logical Operators Short-Circuit:
# If first is False, second is NOT evaluated
if False and expensive_function():  # expensive_function() NOT called
    pass

# If first is True, second is NOT evaluated
if True or expensive_function():  # expensive_function() NOT called
    pass

```python
#H2: Modulo tricks:# Check if a number is:
# 1. Divisible by both 3 and 5 (Divisible by 15)
# 2. Divisible by 3 but NOT by 5
# 3. A perfect square (using %)
# 4. Last digit is 7
# Test all conditions and print results

import math

num = 105

# 1. Divisible by both 3 and 5 (Divisible by 15)
if num % 15 ==0:
    print( "1. Divisible by both 3 and 5 (Divisible by 15)")
else:
    print("1. Not divisible by both 3 and 5")


# 2. Divisible by 3 but NOT by 5
if num % 3 == 0 and num % 5 != 0:
    print("2. Divisible by 3 but NOT by 5")
else:
     print("2. Not divisible by 3 but NOT by 5")

# 3. Perfect square
root = math.isqrt(num)
if root * root == num:
    print("3. Perfect square")
else:
    print("3. Not a perfect square")

# 4. Last digit is 7
if num % 10 == 7:
    print("4. Last digit is 7")
else:
    print("4. Last digit is not 7")

```

12. Cheat Sheet

# ARITHMETIC
+   Addition          a + b
-   Subtraction       a - b
*   Multiplication    a * b
/   Division (float)  a / b
//  Floor Division    a // b
%   Modulo            a % b
**  Power             a ** b

# COMPARISON (Return True/False)
==  Equal to          a == b
!=  Not equal         a != b
>   Greater than      a > b
<   Less than         a < b
>=  Greater or equal  a >= b
<=  Less or equal     a <= b

# LOGICAL
and  Both True        a and b
or   At least 1 True  a or b
not  Reverse boolean  not a

# ASSIGNMENT
=   Assign            a = 5
+=  Add assign        a += 3  (a = a + 3)
-=  Sub assign        a -= 2
*=  Mult assign       a *= 4
/=  Div assign        a /= 2
//= Floor div assign  a //= 3
%=  Mod assign        a %= 2
**= Power assign      a **= 2

# OPERATOR PRECEDENCE (High to Low)
1. **
2. *, /, //, %
3. +, -
4. <, <=, >, >=, ==, !=
5. and
6. or
